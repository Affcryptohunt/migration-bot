```python
from pathlib import Path
import shutil

root = Path('/workspaces/migration-bot/migration-bot')
archive = Path('/workspaces/migration-bot/archive')
archive.mkdir(parents=True, exist_ok=True)

items_to_move = ['api', 'products.csv', 'shopify_products.csv', 'run.sh']

for item in items_to_move:
    src = root / item
    if src.exists():
        dest = archive / item
        shutil.move(str(src), str(dest))
        print(f"Moved {src} -> {dest}")

# Remove generated caches inside the root directory.
for cache_dir in ['__pycache__', 'api/__pycache__']:
    path = root / cache_dir
    if path.exists():
        shutil.rmtree(path)
        print(f"Removed {path}")

print('Reorganized migration-bot directory.')
print('Resulting items:')
for item in sorted(root.iterdir()):
    print(item.name)
```
