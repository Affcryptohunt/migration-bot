```python
from pathlib import Path
import shutil

root = Path('/workspaces/migration-bot/migration-bot')
archive = Path('/workspaces/migration-bot/archive')

# Create new directories
backend = root / 'backend'
storage = root / 'storage'
storage_images = storage / 'images'
storage_exports = storage / 'exports'

backend.mkdir(parents=True, exist_ok=True)
storage_images.mkdir(parents=True, exist_ok=True)
storage_exports.mkdir(parents=True, exist_ok=True)

# List of Python files to move to backend
backend_files = [
    'migrate.py',
    'site_crawler.py',
    'extractor.py',
    'product_detector.py',
    'image_processor.py',
    'csv_exporter.py',
    'platform_detector.py',
    'normalizer.py',
    'category_finder.py'
]

# Move Python files to backend
for file in backend_files:
    src = root / file
    if src.exists():
        dest = backend / file
        shutil.move(str(src), str(dest))
        print(f"Moved {file} -> backend/")

# Move app.py from archive to backend
app_src = archive / 'app.py'
if app_src.exists():
    app_dest = backend / 'app.py'
    shutil.move(str(app_src), str(app_dest))
    print(f"Moved app.py from archive -> backend/")

# Move images from root to storage/images
images_src = root / 'images'
if images_src.exists():
    # Move all image files from images/ to storage/images/
    for image in images_src.iterdir():
        if image.is_file():
            dest = storage_images / image.name
            shutil.move(str(image), str(dest))
    # Remove the now-empty images directory
    images_src.rmdir()
    print(f"Moved images/ to storage/images/")

print("\n✅ Restructuring complete!")
print(f"\nFinal structure in {root}:")
for item in sorted(root.iterdir()):
    if item.name.startswith('.'):
        continue
    if item.is_dir():
        print(f"  {item.name}/")
        for sub in sorted(item.iterdir()):
            if sub.is_dir():
                print(f"    {sub.name}/")
            else:
                print(f"    {sub.name}")
    else:
        print(f"  {item.name}")
```
