```python
from pathlib import Path
import shutil

root = Path('/workspaces/migration-bot/migration-bot')
backend = root / 'backend'
archive = Path('/workspaces/migration-bot/archive')

# 1. Move app.py from archive to backend
app_src = archive / 'api' / 'app.py'
if app_src.exists():
    app_dest = backend / 'app.py'
    shutil.move(str(app_src), str(app_dest))
    print(f"✓ Moved app.py from archive/api/ to backend/")

# 2. Remove category_finder.py from backend (not in requested structure)
category_file = backend / 'category_finder.py'
if category_file.exists():
    category_file.unlink()
    print(f"✓ Removed category_finder.py (not in requested structure)")

print("\n✅ Final Structure:")
print(f"\n{root}/")
print("├── backend/")
for f in sorted(backend.glob("*.py")):
    print(f"│   ├── {f.name}")
print("│")
print("├── frontend/")
for f in (root / 'frontend').glob("*"):
    print(f"│   └── {f.name}")
print("│")
print("├── storage/")
storage = root / 'storage'
for sub in sorted(storage.glob("*")):
    if sub.is_dir():
        print(f"│   ├── {sub.name}/")
print("│")
print("├── requirements.txt")
print("└── README.md")

# Verify structure matches exactly
backend_files = sorted([f.name for f in backend.glob("*.py")])
expected = ['app.py', 'csv_exporter.py', 'extractor.py', 'image_processor.py', 
            'migrate.py', 'normalizer.py', 'platform_detector.py', 
            'product_detector.py', 'site_crawler.py']

print(f"\n✅ Backend files match expected structure: {backend_files == expected}")
```
